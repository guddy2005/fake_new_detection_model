```python
import pandas as pd
import re

# Load datasets
df_fake = pd.read_csv("/content/Fake.csv")
df_true = pd.read_csv("/content/True.csv")

# Add labels
df_fake['label'] = 0
df_true['label'] = 1

# Combine
df_global = pd.concat([df_fake, df_true], ignore_index=True)

# Keep only required columns
df_global = df_global[['text','label']]

# 🔥 Remove Reuters & location bias
def clean_news(text):
    text = str(text)
    text = re.sub(r'\(Reuters\)', '', text, flags=re.IGNORECASE)
    text = re.sub(r'^[A-Z\s]+\s-\s', '', text)  # Remove WASHINGTON -
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df_global['text'] = df_global['text'].apply(clean_news)

# Remove null & duplicates
df_global = df_global.dropna().drop_duplicates()

# Save
df_global.to_csv("/content/global_cleaned.csv", index=False)

print("Final shape:", df_global.shape)
```

```python
import pandas as pd

df_indian = pd.read_csv("/content/indian_cleaned.csv")
df_global = pd.read_csv("/content/global_cleaned.csv")

# Take 10k from Indian
df_indian_sample = df_indian.sample(n=10000, random_state=42)

# Take 10k from Global
df_global_sample = df_global.sample(n=10000, random_state=42)

df_final = pd.concat([df_indian_sample, df_global_sample], ignore_index=True)

df_final = df_final.sample(frac=1, random_state=42).reset_index(drop=True)

print("Final shape:", df_final.shape)

df_final.to_csv("/content/final_dataset.csv", index=False)
```
